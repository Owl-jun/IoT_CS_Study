### 네트워크란 무엇일까

- 네트워크
    - 컴퓨터가 탑재된 장치(컴퓨터,스마트폰,서버 등) 이 데이터를 주고받는 것이다.
    - 네트워크가 없다면, 인터넷은 존재하지 않을 것이고, 파일 공유 및 서버 연결도 되지 않는다.

- 기본개념 클라이언트 & 서버
    - 클라이언트 : 요청을 보내는 쪽 (ex: 웹브라우저)
    - 서버 : 요청을 받아서 응답하는 쪽 (ex: 웹사이트)

- 내가 웹사이트를 방문 할 때 무슨 일이 일어날까?
    - 나의 생각 
        - DNS 서버에 웹사이트 도메인 주소의 IP주소를 요청한다.
        - IP주소를 받은 후, HTTP요청을 서버에 보낸다.
        - 서버는 요청을 확인하고, 관련 데이터를 전송한다.
        - 웹브라우저는 데이터를 조립해서 화면에 표시한다.

### IP 주소와 DNS (인터넷 주소의 원리)

- 컴퓨터들은 서로를 어떻게 찾아낼까?

- IP 주소
    - 네트워크 상에서 각 기기를 구별하는 "주소
        - ex: 192.168.0.1 (IPv4) , 2001:db8::1 (IPv6)
        - 추후 IPv4와 IPv6 의 특징과 차이점을 공부하자
    
    - 공인 IP vs 사설 IP
        - 공인 IP : 전 세계에서 접근 가능하며 **ISP(인터넷 서비스 제공자)**가 할당한 주소
        - 사설 IP : 내부 네트워크 전용 주소 (라우터를 통해 공인IP로 변환된다.)
        - NAT(Network Address Translation) : 사설 IP를 공인 IP로 변환하는 기술
    
    - DNS(Domain Name System)
        - IP주소를 사람이 읽기 쉽게 바꿔주는 것이다.
        - www.google.com -> DNS -> 142.250.74.206 (반대로도 가능)

    - 실습
        - Win + R 키 눌러 실행창 띄우기
        - cmd 입력 후 엔터
        - 터미널에 ping google.com 실행해보기
        - nslookup google.com 혹은 dig google.com 실행해서 DNS 변환 확인

### TCP/IP 모델 & OSI 7계층

- 나중에 **TCP** 좀더 깊이 들어가 볼 것.

- TCP/IP 모델 (4계층)
    - 네트워크 인터페이스 (Link Layer) : 물리적인 연결 (LAN, Wi-Fi)
    - 인터넷 계층 (IP Layer) : 목적지 찾기 (IP주소)
    - 전송 계층 (Transport Layer) : 신뢰성 보장 (TCP/UDP)
    - 응용 계층 (Application Layer) : HTTP,FTP,SMTP 등

- OSI 7계층
    - 물리 계층 (Physical Layer) → 신호 전송 (LAN 케이블, Wi-Fi)
    - 데이터 링크 계층 (Data Link Layer) → MAC 주소 기반 데이터 전달 (이더넷, 스위치)
    - 네트워크 계층 (Network Layer) → IP 주소 기반 패킷 전송 (IP, 라우터)
    - 전송 계층 (Transport Layer) → 신뢰성 있는 데이터 전달 (TCP/UDP)
    - 세션 계층 (Session Layer) → 연결 세션 관리 (SSL/TLS)
    - 표현 계층 (Presentation Layer) → 데이터 암호화 & 변환 (JPEG, GIF, TLS)
    - 응용 계층 (Application Layer) → 최종 사용자 인터페이스 (HTTP, FTP, SMTP)

### TCP와 UDP의 차이
- 인터넷에서 데이터가 오가는 방식 이해하기

- TCP (Transmission Control Protocol)

    - 특징: **연결 지향**, **신뢰성 보장** (패킷 손실 시 재전송)
    - 예: 웹브라우저(HTTP/HTTPS), 이메일(SMTP), 파일 전송(FTP)

- UDP (User Datagram Protocol)

    - 특징: **연결 없이 빠름**, 신뢰성 낮음
    - 예: 실시간 스트리밍(YouTube), 온라인 게임, VoIP(카카오톡 통화)

- 실습:
    - netstat -an 명령어 실행해서 현재 TCP/UDP 연결 확인하기
    - Wireshark로 패킷 캡처해보기 (심화 학습)

```python
!netstat -an
```

### 라우터와 스위치의 차이
- 네트워크 장비의 역할 이해하기
    
    - IP : 네트워크 상에서 나의 주소
    - MAC : 같은 네트워크 안 에서, 각 하드웨어 주소
    - 라우터(Router) : 네트워크 간 데이터 전달 (IP 기반)
    - 스위치(Switch) : 같은 네트워크 내에서 데이터 전달 (MAC 주소 기반)

### 네트워크 보안 기초 (기본적인 공격과 방어)
- 네트워크 보안 개념 익히기

    - 패킷 스니핑 (Packet Sniffing) → 네트워크 트래픽 감청 (Wireshark 활용)
    - DDoS (Distributed Denial of Service) 공격 → 서버에 과부하 유발
    - 방화벽 (Firewall) → 허용된 트래픽만 통과시키고, 불법 접근 차단
    - VPN (Virtual Private Network) → 암호화된 터널을 통해 안전한 인터넷 사용
    - MITM (Man-In-The-Middle) 공격 → 공격자가 중간에서 통신 가로채기
    - ARP 스푸핑 → 공격자가 네트워크에서 위장하여 데이터 가로채기

    
- 실습:
    - iptables (Linux) 또는 Windows Defender Firewall 설정 확인
